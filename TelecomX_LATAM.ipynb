```python
import requests
import pandas as pd

url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"

response = requests.get(url)
```

```python
data_json = response.json()
telecomx = pd.DataFrame(data_json)
telecomx.head()
```

```python
telecomx_expandido = pd.concat([
    telecomx.drop(['customer', 'phone', 'internet', 'account'], axis=1),
    telecomx['customer'].apply(pd.Series),
    telecomx['phone'].apply(pd.Series),
    telecomx['internet'].apply(pd.Series),
    telecomx['account'].apply(pd.Series)
], axis=1)
```

```python
telecomx_expandido
```

```python
charges_df = telecomx_expandido['Charges'].apply(pd.Series)

charges_df['Total'] = pd.to_numeric(charges_df['Total'], errors='coerce')

telecomx_expandido = pd.concat([telecomx_expandido.drop(columns=['Charges']), charges_df.rename(columns={'Monthly': 'MonthlyCharges', 'Total': 'TotalCharges'})], axis=1)
```

```python
telecomx_expandido
```

```python
print(telecomx_expandido.columns.tolist())
```

```python
columnas_claves = [
    "Contract",
    "tenure",
    "MonthlyCharges",
    "PaymentMethod",
    "OnlineSecurity",
    "TechSupport"
]

telecomx_claves = telecomx_expandido[columnas_claves]

telecomx_claves.head()
```

```python
print(telecomx_expandido.dtypes)
```

```python
for col in telecomx_expandido.select_dtypes(include='object').columns:
    print(f"\n {col} â†’ {telecomx_expandido[col].unique()}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Reemplazar string vacÃ­o por NaN
telecomx_expandido['Churn'] = telecomx_expandido['Churn'].replace('', np.nan)

# Verificar cuÃ¡ntos quedan nulos
print(telecomx_expandido['Churn'].isnull().sum())
```

```python
telecomx_expandido
```

### ComprobaciÃ³n de incoherencias en los datos

```python
binarias = [
    'Churn', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Reemplazo binario
telecomx_expandido[binarias] = telecomx_expandido[binarias].replace({'Yes': 1, 'No': 0})

telecomx_expandido[binarias] = telecomx_expandido[binarias].replace({'Yes': 1, 'No': 0})
telecomx_expandido[binarias] = telecomx_expandido[binarias].infer_objects(copy=False)

telecomx_expandido[binarias]
```

```python
telecomx_expandido
```

### Crear la columna "Cuentas_Diarias"

```python
telecomx_expandido["Cuentas_Diarias"] = telecomx_expandido["MonthlyCharges"] / 30
telecomx_expandido
```

```python
# GrÃ¡fico de barras
telecomx_expandido["Churn_vis"] = telecomx_expandido["Churn"].map({1: "Yes", 0: "No"})

churn_df = telecomx_expandido["Churn_vis"].value_counts().reset_index()
churn_df.columns = ["Churn", "Cantidad"]

plt.figure(figsize=(6, 4))
sns.barplot(data=churn_df, x="Churn", y="Cantidad", hue="Churn",
            palette={"No": "#4CAF50", "Yes": "#F44336"}, legend=False)

plt.title("DistribuciÃ³n de EvasiÃ³n (Churn)")
plt.xlabel("Estado del Cliente")
plt.ylabel("Cantidad de Clientes")
plt.tight_layout()

plt.savefig(f'{ruta_graficos}/distribucion_evasion.png')

plt.show()
```

```python

# GrÃ¡fico de pastel

churn_counts = telecomx_expandido["Churn"].value_counts()

labels = ["No (Activo)", "Yes (EvasiÃ³n)"]
colors = ["#4CAF50", "#F44336"]

plt.figure(figsize=(5, 5))
plt.pie(churn_counts.values, labels=labels, autopct="%1.1f%%",
        startangle=90, colors=colors, wedgeprops={"edgecolor": "white"})
plt.title("ProporciÃ³n de Clientes con EvasiÃ³n")
plt.axis("equal")  # Asegura que el cÃ­rculo sea perfecto

plt.savefig(f'{ruta_graficos}/clientes_evasion.png')

plt.show()
```

# **Conclusiones e Insights**
- Los clientes que evaden tienden a tener menor tiempo de contrato y menor gasto total acumulado.
- El mÃ©todo de pago "Electronic check" muestra una mayor proporciÃ³n de evasiÃ³n comparado con otros mÃ©todos.
- Los contratos "Month-to-month" estÃ¡n fuertemente asociados con churn, mientras que "Two year" presentan mayor retenciÃ³n.
- No se observan diferencias significativas en churn por gÃ©nero.

# **Recomendaciones EstratÃ©gicas**

- Incentivar contratos a largo plazo mediante beneficios exclusivos para clientes con "One year" o "Two year".
- Monitorear clientes con bajo tenure y bajo gasto para aplicar campaÃ±as de retenciÃ³n temprana.
- Revisar la experiencia de pago con "Electronic check", ya que podrÃ­a estar asociada a fricciÃ³n o insatisfacciÃ³n.
- Segmentar clientes por perfil de riesgo y aplicar modelos predictivos para anticipar evasiÃ³n.
